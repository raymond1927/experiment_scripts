```python
import matplotlib
import csv
import matplotlib.pyplot as plt

font_size = {'size': '14'}
```

```python
air_total_energy_deposited = []
air_kinetic_energy_difference = []

spiro_total_energy_deposited = []
spiro_kinetic_energy_difference = []

pero_total_energy_deposited = []
pero_kinetic_energy_difference = []

# files without the gold buffer layer have the prefix no
# files with the au prefix have a gold layer
# even though there are two files only the first one is required as the second file is concatenated
# to the first file
for eng_level in range(5,16):
    with open('no' + str(eng_level) + 'kt0.csv') as csvfile:
        reader = csv.reader(csvfile)
        
        air_total_energy = []
        air_kinetic_energy = []
        
        spiro_total_energy = []
        spiro_kinetic_energy = []

        pero_total_energy = []
        pero_kinetic_energy = []

        for row in reader:
            if len(row) > 1:
                if float(row[4]) < -420.0:
                    air_total_energy.append(float(row[7]))
                    air_kinetic_energy.append(float(row[5]))
                elif float(row[4]) < -270.0:
                    spiro_total_energy.append(float(row[7]))
                    spiro_kinetic_energy.append(float(row[5]))
                elif float(row[4]) < 130.0:
                    pero_total_energy.append(float(row[7]))
                    pero_kinetic_energy.append(float(row[5])) 
                
        air_total_energy_deposited.append(
            sum(air_total_energy) if air_total_energy else 0)
        air_kinetic_energy_difference.append(
            sum(air_kinetic_energy) if air_kinetic_energy else 0)

        spiro_total_energy_deposited.append(
            sum(spiro_total_energy) if spiro_total_energy else 0)
        spiro_kinetic_energy_difference.append(
            sum(spiro_kinetic_energy) if spiro_kinetic_energy else 0)

        pero_total_energy_deposited.append(
            sum(pero_total_energy) if pero_total_energy else 0)
        pero_kinetic_energy_difference.append(
            sum(pero_kinetic_energy) if pero_kinetic_energy else 0)
```

```python
# Plotting the total energy deposited

mev_air_total_energy = [i/1000000 for i in air_total_energy_deposited]
mev_spiro_total_energy = [i/1000000 for i in spiro_total_energy_deposited]
mev_pero_total_energy = [i/1000000 for i in pero_total_energy_deposited]

energies = range(5,16)
matplotlib.rc('font', **font_size)
plt.plot(energies, mev_air_total_energy, label='Air')
plt.plot(energies, mev_spiro_total_energy, label='Spiro-OMeTAD')
plt.plot(energies, mev_pero_total_energy, label='Perovskite')
plt.title('Total energy deposited for range of energy with no gold layer')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Total energy deposited (MeV)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the total kinetic energy deposited at each step

matplotlib.rc('font', **font_size)
energies = range(5,16)
plt.close()
plt.plot(energies, [i/1000000 for i in air_kinetic_energy_difference], label='Air')
plt.plot(energies, [i/1000000 for i in spiro_kinetic_energy_difference], label='Spiro-OMeTAD')
plt.plot(energies, [i/1000000 for i in pero_kinetic_energy_difference], label='Perovskite')
plt.title('Total kinetic energy difference for range of energy with no gold layer')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Total energy deposited (MeV)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

au_au_total_energy_deposited = []
au_au_kinetic_energy_difference = []

au_spiro_total_energy_deposited = []
au_spiro_kinetic_energy_difference = []

au_pero_total_energy_deposited = []
au_pero_kinetic_energy_difference = []

# plot for all the data with the gold buffer layer
for eng_level in range(5,16):
    with open('au' + str(eng_level) + 'kt0.csv') as csvfile:
        reader = csv.reader(csvfile)
        
        au_total_energy = []
        au_kinetic_energy = []
        
        spiro_total_energy = []
        spiro_kinetic_energy = []

        pero_total_energy = []
        pero_kinetic_energy = []

        for row in reader:
            if len(row) > 1:
                if float(row[4]) < -420:
                    au_total_energy.append(float(row[7]))
                    au_kinetic_energy.append(float(row[5]))
                elif float(row[4]) < -270:
                    spiro_total_energy.append(float(row[7]))
                    spiro_kinetic_energy.append(float(row[5]))
                elif float(row[4]) < 130:
                    pero_total_energy.append(float(row[7]))
                    pero_kinetic_energy.append(float(row[5])) 
                
        au_au_total_energy_deposited.append(
            sum(au_total_energy) if au_total_energy else 0)
        au_au_kinetic_energy_difference.append(
            sum(au_kinetic_energy) if au_kinetic_energy else 0)

        au_spiro_total_energy_deposited.append(
            sum(spiro_total_energy) if spiro_total_energy else 0)
        au_spiro_kinetic_energy_difference.append(
            sum(spiro_kinetic_energy) if spiro_kinetic_energy else 0)

        au_pero_total_energy_deposited.append(
            sum(pero_total_energy) if pero_total_energy else 0)
        au_pero_kinetic_energy_difference.append(
            sum(pero_kinetic_energy) if pero_kinetic_energy else 0)
```

```python
# Plotting the total energy deposited


matplotlib.rc('font', **font_size)
energies = range(5,16)
plt.plot(energies, [i/1000000 for i in au_au_total_energy_deposited], label='Gold')
plt.plot(energies, [i/1000000 for i in au_spiro_total_energy_deposited], label='Spiro-OMeTAD')
plt.plot(energies, [i/1000000 for i in au_pero_total_energy_deposited], label='Perovskite')
plt.title('Total energy deposited for range of energy with gold layer')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Total energy deposited (MeV)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting the total kinetic energy deposited at each step

matplotlib.rc('font', **font_size)
energies = range(5,16)
plt.close()
plt.plot(energies, [i/1000000 for i in au_au_kinetic_energy_difference], label='Gold')
plt.plot(energies, [i/1000000 for i in au_spiro_kinetic_energy_difference], label='Spiro-OMeTAD')
plt.plot(energies, [i/1000000 for i in au_pero_kinetic_energy_difference], label='Perovskite')
plt.title('Total kinetic energy difference for range of energy with gold layer')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Total energy deposited (MeV)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Comparing the energy of gold vs air layer

matplotlib.rc('font', **font_size)
energies = range(5,16)
plt.close()
plt.plot(energies, [i/1000000 for i in air_kinetic_energy_difference], label='No gold')
plt.plot(energies, [i/1000000 for i in au_au_kinetic_energy_difference], label='Gold')
plt.title('Total kinetic energy deposited into gold layer')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Total energy deposited (MeV)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Comparing the energy of gold vs air layer

matplotlib.rc('font', **font_size)
energies = range(5,16)
plt.close()
plt.plot(energies, [i/1000000 for i in pero_kinetic_energy_difference], label='No gold')
plt.plot(energies, [i/1000000 for i in au_pero_kinetic_energy_difference], label='Gold')
plt.title('Total kinetic energy deposited into Perovskite layer')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Total energy deposited (MeV)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import csv

energies = range(5,16)
with open('perovskite_kinetic_energy.csv', mode='w') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(['Electron energy (keV)', 'No Au (eV)', 'Au (eV)'])
    for i in range(len(energies)):
        csv_writer.writerow([energies[i], pero_kinetic_energy_difference[i], au_pero_kinetic_energy_difference[i]])
```

```python
# Checking total energy of system
matplotlib.rc('font', **font_size)
energies = range(5,16)
diff_in_kinetic_energy = []
for i in range(len(energies)):
    au_total_kinetic = au_au_kinetic_energy_difference[i] + au_spiro_kinetic_energy_difference[i] +  au_pero_kinetic_energy_difference[i]
    
    no_total_kinetic = air_kinetic_energy_difference[i] + spiro_kinetic_energy_difference[i] + pero_kinetic_energy_difference[i]

    diff_in_kinetic_energy.append(no_total_kinetic - au_total_kinetic)

plt.close()
plt.plot(energies, [i/1000000 for i in diff_in_kinetic_energy])
plt.title('Diff. in total kinetic energy deposited\n from gold/no gold layer BV')
plt.xlabel('Electron energy (keV)')
plt.ylabel('Diff. in total energy\n deposited (MeV)')
plt.grid(True)
plt.show()
```
