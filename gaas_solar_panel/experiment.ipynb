```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from tabulate import tabulate

# energy of electron beam
energy = [10, 15, 20, 25, 30]

# current of elctron beam
current = [4, 5, 4.98, 5.02, 5]

# current generated by solar panel
collection = [0.22, 1.61, 2.16, 2.67, 3.03]

# current calculated by simulation
simulation = [0.78, 2.05, 1.85, 1.82, 1.19]

input_power = [a * b for a, b in zip(energy, current)]
output_power = [a * 1.8 for a in collection]
eff = [a * 100 / b for a, b in zip(output_power, input_power)]

energy_copy = energy.copy()
input_power_copy = input_power.copy()
output_power_copy = output_power.copy()
eff_copy = eff.copy()

energy_copy.insert(0, 'Energy (keV)')
input_power_copy.insert(0, 'Input power')
output_power_copy.insert(0, 'Output power')
eff_copy.insert(0, 'Efficiency (%)')

print(tabulate([energy_copy, input_power_copy, output_power_copy, eff_copy]))

font_size = {'size': 14}
matplotlib.rc('font', **font_size)

plt.plot(energy, collection, linestyle='-', color='b', label='Experiment')
plt.plot(energy, simulation, color='red', label='Simulation')
plt.title("Current generated at different electron energy")
plt.xlabel("Energy (keV)")
plt.ylabel("Current (uA)")
plt.grid(True)
plt.legend()
plt.show()
```
