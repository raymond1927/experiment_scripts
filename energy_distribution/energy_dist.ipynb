```python
# read all the data from the file
import csv
import re

radioisotopes = {}

with open('ICRP-07.BET') as csvfile:
    reader = csv.reader(csvfile)
    current_radioisotope = None
    for row in reader:
        cols = row[0].split()
        first = cols[0].strip()
        second = cols[1].strip()
        if re.search('[a-zA-Z]', first):
            current_radioisotope = first
            radioisotopes[current_radioisotope] = {}
            radioisotopes[current_radioisotope]['probability'] = []
            radioisotopes[current_radioisotope]['energy'] = []
        else:
            radioisotopes[current_radioisotope]['energy'].append(float(first))
            radioisotopes[current_radioisotope]['probability'].append(float(second))
```

```python
# print out all radioisotopes
for key in sorted(radioisotopes.keys()):
    print(key)
```

```python
# visualise the distribution of a choosen radioisotope
import matplotlib.pyplot as plt
import matplotlib

# set large font for matlab
font_size = {'size': '14'}
matplotlib.rc('font', **font_size)

radioisotope = input('Enter the name of the radioisotope: ')
if radioisotope in radioisotopes:
    probability = radioisotopes[radioisotope]['probability']
    energy = radioisotopes[radioisotope]['energy']

    total_probability = sum(probability)
    total_length = 100000
    cumulative_probability = []
    current_probability = 0

    for i in probability:
        current_probability += i
        cumulative_probability.append(current_probability/total_probability)

    plt.plot(cumulative_probability, energy)
    plt.grid(True)
    plt.xlabel('Cumulative probability')
    plt.ylabel('Energy (MeV)')
    plt.title('Cumulative energy distribution for ' + radioisotope)
    plt.show()

else:
    print('Radioisotope ' + radioisotope + ' not found!')
```
